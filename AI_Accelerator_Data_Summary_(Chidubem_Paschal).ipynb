The purpose of this python coding is to give an overview of the dataset introduced here and its mathematical summary

```python
#let's import our python packages
import pandas as pd
import numpy as np
```

```python
#let's import the dataset we're working with
My_data = pd.read_csv('https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv')
My_data.head()
```

```python
#Here is the info of my whole dataset
My_data.info()
```

```python
#Now the mathematical summary
My_data.describe()
```

```python
#I will now get the median and mode of the data
My_data.median()
'/n'
My_data.mode()
```

```python
#generating the dataset for values when the median_income is greater than the median
Topdata = My_data[My_data['median_income'] > 3.125]
Topdata
```

```python
#generating the dataset for values when the median_income is lesser than the median
Bottomdata = My_data[My_data['median_income'] <= 3.125]
Bottomdata
```
